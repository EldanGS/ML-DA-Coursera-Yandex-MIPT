```python

import scipy.linalg
from numpy import sin, exp

f_x = lambda x: sin(x / 5.) * exp(x / 10.) + 5. * exp(-x / 2.) # f(x) 

p_rank = 1 #   
A_1 = [
    [1 ** n for n in range(0, p_rank + 1)],
    [15 ** n for n in range(0, p_rank + 1)]
]

b_1 = [f_x(1), f_x(15)]
scipy.linalg.solve(A_1, b_1)

p_rank = 2
A_2 = [
    [1 ** n for n in range(0, p_rank + 1)],
    [8 ** n for n in range(0, p_rank + 1)],
    [15 ** n for n in range(0, p_rank + 1)]
]

b_2 = [f_x(1), f_x(8), f_x(15)]
scipy.linalg.solve(A_2, b_2)

p_rank = 3
A_3 = [
    [1 ** n for n in range(0, p_rank + 1)],
    [4 ** n for n in range(0, p_rank + 1)],
    [10 ** n for n in range(0, p_rank + 1)],
    [15 ** n for n in range(0, p_rank + 1)]
]

b_3 = [f_x(1), f_x(4), f_x(10), f_x(15)]
answer = scipy.linalg.solve(A_3, b_3)

result = [round(x, 2) for x in answer]
result
```
