```python
%matplotlib inline

from sklearn import datasets, cross_validation, naive_bayes
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data_digits = datasets.load_digits()
data_cancer =_cancer = datasets.load_breast_cancer()
```

```python
data_digits.keys()
```

```python
data_cancer.keys()
```

```python
data_cancer['DESCR']
```

```python
cancer_X = data_cancer['data']
cancer_y = data_cancer['target']
```

```python
digits_X = data_digits['data']
digits_y = data_digits['target']
```

```python
for col in pd.DataFrame(digits_X).columns[:5]:
    pd.DataFrame(digits_X)[[col]].hist()
```

```python
for col in pd.DataFrame(cancer_X).columns[:5]:
    pd.DataFrame(cancer_X)[[col]].hist()
```

```python
clf_types = {
    'bernoulli': naive_bayes.BernoulliNB(),
    'multinomial': naive_bayes.MultinomialNB(),
    'gaussian': naive_bayes.GaussianNB()
}
```

```python
cancer_scores = {}
for clf_type in clf_types:
    cancer_scores[clf_type] = cross_validation.cross_val_score(clf_types[clf_type], cancer_X, cancer_y).mean()

print(cancer_scores)
```

```python
digits_scores = {}
for clf_type in clf_types:
    digits_scores[clf_type] = cross_validation.cross_val_score(clf_types[clf_type], digits_X, digits_y).mean()

print(digits_scores)
```
